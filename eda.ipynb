```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=['user_id', 'movie_id', 'rating', 'timestamp'])
ratings.head()
```

```python
info = pd.read_csv('ml-100k/u.info', sep=' ', names=['count', 'type'])
info
```

```python
movie_genres = pd.read_csv('ml-100k/u.genre', sep='|', names=['genre', 'genre_id'])
movie_genres
```

```python
genres = list(movie_genres['genre'])
columns = ['movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL'] + genres
items = pd.read_csv('ml-100k/u.item', sep='|',header=None,names=columns,encoding='latin-1')
items
```

```python
occupation = pd.read_csv('ml-100k/u.occupation', names=['occupation'])
occupation
```

```python
user = pd.read_csv('ml-100k/u.user', sep='|', names=['age', 'sex', 'occupation', 'zip_code'])
user
```

```python
plt.hist(ratings['rating'], bins=5, edgecolor='black')
plt.title("Distribution of Ratings for Movies")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
ratings_with_titles = pd.merge(ratings, items[['movie_id','title']], on='movie_id')

rating_counts = ratings_with_titles.groupby('title').size().sort_values(ascending=False)
top_10 = rating_counts.head(10)
print("Top 10 Most Rated Movies:")
print(top_10)

plt.figure(figsize=(12, 6))
plt.bar(range(len(rating_counts)), rating_counts, color='skyblue', edgecolor='black')
plt.title("Number of Ratings per Movie")
plt.xlabel("Movie Index")
plt.ylabel("Number of Ratings")
plt.xticks(range(0, len(rating_counts), 200), range(1, len(rating_counts) + 1, 200), rotation=0)
plt.tight_layout()
plt.show()
```

```python
avg_ratings = ratings_with_titles.groupby('title')['rating'].mean()
top_20_avg = avg_ratings.sort_values(ascending=False).head(20)
print("Top 20 Highest Average Rating:")
print(top_20_avg)
```

```python
ratings_per_user = ratings.groupby('user_id').size()
print("Min ratings by a user:", ratings_per_user.min())
print("Max ratings by a user:", ratings_per_user.max())

plt.hist(ratings_per_user, bins=1682, edgecolor='black')
plt.title("Distribution of Number of Ratings per User")
plt.xlabel("num of Ratings")
plt.ylabel("num of Users")
plt.show()
```
